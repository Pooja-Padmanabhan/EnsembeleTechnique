#### Step 1: Import all the required libraries

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pickle
```

#### Step 2: Import the csv file.

```python

def load_data(merged_df):
    """Load data from CSV file"""
    merged_df = pd.read_csv('C:/Users/Pooja/Downloads/merged_df.csv')
    return merged_df
```

#### Step 3: Perform EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns

def perform_eda(merged_df):
    """
    Performs Exploratory Data Analysis (EDA) on a Pandas DataFrame.
    
    Args:
    df (pandas.DataFrame): The DataFrame to perform EDA on.
    
    Returns:
    None.
    """
    # Print the first five rows of the DataFrame
    print("First five rows of the DataFrame:")
    print(merged_df.head())

    # Print the last five rows of the DataFrame
    print("Last five rows of the DataFrame:")
    print(merged_df.tail())

    # Print a summary of the DataFrame
    print("Summary of the DataFrame:")
    print(merged_df.describe())

    # Print information about the DataFrame
    print("Information about the DataFrame:")
    print(merged_df.info())

    # Check for missing values in the DataFrame
    print("Number of missing values in the DataFrame:")
    print(merged_df.isnull().sum())

    # Plot a histogram of the numerical columns
    print("Histogram of the numerical columns:")
    merged_df.hist(figsize=(15, 15))
    plt.show()

    # Plot a correlation matrix of the numerical columns
    fig, ax = plt.subplots(figsize=(19,19))
    print("Correlation matrix of the numerical columns:")
    sns.heatmap(merged_df.corr(),annot=True,cmap='magma',square=True,linewidths=0.3,linecolor='black')
    plt.show()
```

#### Step 4: Preprocess the data by using the appropriate standardisation/normalisation.

```python
def preprocess_data(merged_df):
    """Preprocess data and split into features and target"""
    x = merged_df.drop('Churn', axis=1)
    y = merged_df['Churn']
    return x, y
```

#### Step 5: Split the dataset into training and testing sets

```python
def split_data(x, y, test_size=0.2, random_state=42):
    """Split data into train and test sets"""
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state=random_state)
    return x_train, x_test, y_train, y_test
```

#### Step 6: Standardize the test and train data

```python
def scale_data(x_train, x_test):
    """Scale data using standard scaler"""
    scaler = StandardScaler()
    x_train = scaler.fit_transform(x_train)
    x_test = scaler.transform(x_test)
    return x_train, x_test
```

#### Step 7: Train various base models

```python
def train_model(x_train, y_train, model_type='lr'):
    """Train a machine learning model"""
    if model_type == 'lr':
        model = LogisticRegression()
    elif model_type == 'rf':
        model = RandomForestClassifier()
        
    elif model_type == 'dt':
        model = DecisionTreeClassifier()
    elif model_type == 'xgb':
        model = XGBClassifier()
    else:
        raise ValueError('Invalid model type')
        
    model.fit(x_train, y_train)
    return model
```

#### Step 8: Evaluate the performance of the models and select the best performing model

```python
def evaluate_model(model, x_test, y_test):
    """Evaluate the performance of a machine learning model"""
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return accuracy, precision, recall, f1
```

#### Step 9: Save the best performing model as a pickle file

```python
def save_model(model, model_filename='model.pkl'):
    """Save a machine learning model to a file"""
    with open(model_filename, 'wb') as f:
        pickle.dump(model, f)
```

#### Step 10: Put all the functions together in a main function to build the ML workflow:

```python
def main(merged_df_file, model_type='lr', test_size=0.2, random_state=42, model_filename='model.pkl'):
    """Main function that executes the entire ML workflow"""
    merged_df = load_data(merged_df_file)
    x, y = preprocess_data(merged_df)
    x_train, x_test, y_train, y_test = split_data(x, y, test_size=test_size, random_state=random_state)
    x_train, x_test = scale_data(x_train, x_test)
    model = train_model(x_train, y_train, model_type=model_type)
    accuracy, precision, recall, f1 = evaluate_model(model, x_test, y_test)
    print(f"Accuracy: {accuracy:.3f}")
    print(f"Precision: {precision:.3f}")
    print(f"Recall: {recall:.3f}")
    print(f"F1 score: {f1:.3f}")
    save_model(model, model_filename=model_filename)
    return {'accuracy': accuracy, 'precision': precision, 'recall': recall, 'f1': f1}
```

#### Step 11: Best model

```python
def train_best_model(merged_df_file, model_type='lr', model_filename='best_model.pkl'):
    """Train and save the best performing model"""
    merged_df = load_data(merged_df_file)
    x, y = preprocess_data(merged_df)
    x_train, x_test, y_train, y_test = split_data(x, y, test_size=0.2, random_state=42)
    x_train, x_test = scale_data(x_train, x_test)
    best_model = train_model(x_train, y_train, model_type=model_type)
    save_model(best_model, model_filename=model_filename)

if __name__ == '__main__':
    merged_df_file = 'C:/Users/Admin/Downloads/merged_df.csv'
    perform_eda(merged_df)
    performance_df = {}
    for model_type in ['lr','rf', 'dt', 'xgb']:
        print(f"Model Type: {model_type}")
        performance = main(merged_df_file, model_type=model_type, test_size=0.2, random_state=42, model_filename=f"{model_type}_model.pkl")
        performance_df[model_type] = performance
        print("\n")
    
    # Print model performance
    best_model_name = max(performance_df, key=lambda key: performance_df[key]['accuracy'])
    best_model_performance = performance_df[best_model_name]
    print(f"Best model: {best_model_name}")
    print(f"Accuracy: {best_model_performance['accuracy']:.4f}")
    print(f"Precision: {best_model_performance['precision']:.4f}")
    print(f"Recall: {best_model_performance['recall']:.4f}")
    print(f"F1 score: {best_model_performance['f1']:.4f}")
    
    # Train and save the best model
    train_best_model(merged_df_file, model_type=best_model_name, model_filename='best_model.pkl')
```
